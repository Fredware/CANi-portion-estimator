# EEG Data Analysis with MNE-Python 

##  Libraries and Notebook Setup

```python
import mne
import numpy as np
import scipy.io

print(mne.__version__)
```

```python
%matplotlib qt
```

## Loading MATLAB File

```python
matlab_data = scipy.io.loadmat('NH_test1.mat')
print(matlab_data.keys())
```

## Creating MNE-Python Data Structures from Scratch

### Create the `Info` object
To initialize a minimal Info object requires a list of channel names, and the sampling frequency.

```python
matlab_ch_labels = matlab_data['electrode_labels'].squeeze().tolist()
ch_names = [label[0] for label in matlab_ch_labels]

ch_types = ['eeg']*len(ch_names) # optional

fs = matlab_data['fs']

info = mne.create_info(ch_names, ch_types=ch_types, sfreq=fs)
info.set_montage('standard_1020')
print(info)
```

### Create the `Raw` Object

Use the `mne.io.RawArray` class constructor, which takes an `Info` object and a `NumPy array` of shape `(n_channels, n_samples)` 

```python
raw = mne.io.RawArray(matlab_data['eeglab_version'], info)
```

```python
raw.plot(scalings=15)
```

```python
raw.plot_psd()
```

### Add Events and Annotations

```python
info = mne.create_info(['STI'], raw.info['sfreq'], ['stim'])
stim_raw = mne.io.RawArray(matlab_data['evnts'], info)
raw.add_channels([stim_raw], force_update_info=True)
```

```python
event_desc = {'Up':10,
              'Down':20,
              'Left':30, 
              'Right':40, 
              'Blank':80, 
              'Cross':90, 
              'Instructions':250, 
              'Start':255, 
              'Unknown':1}

mne_events = mne.find_events(raw, stim_channel='STI')

mne.viz.plot_events(mne_events, sfreq=raw.info['sfreq'], first_samp=raw.first_samp, event_id=event_desc)
```

```python
annot_desc = {event_desc[i]:i for i in event_desc}

annotations = mne.annotations_from_events(
    events=mne_events,
    event_desc = annot_desc,
    sfreq = raw.info['sfreq']
)

raw.set_annotations(annotations)
```

```python
raw.plot(scalings=15)
```
